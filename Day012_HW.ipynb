## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
from collections import defaultdict

news_by_class = defaultdict(list)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all('a')[-1].text)
    news_by_class[d.find(class_="tag").text].append(d)
```

### ③ 哪一個類別的文章最多

```python
max_cnt = 0
max_cnt_class = ""
for key in news_by_class.keys():
    if len(news_by_class[key]) > max_cnt:
        max_cnt = len(news_by_class[key])
        max_cnt_class = key
        
print("數量最多的新聞類別是", max_cnt_class, "共有", max_cnt, "篇")
```
